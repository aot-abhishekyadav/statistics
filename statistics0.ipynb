```python
# method for finding outliers
```

```python
# creating Dataset
dataset = [11, 10, 12, 14, 12, 15, 14, 13, 15, 102, 12, 14, 17, 19, 107, 10, 13, 12, 14, 12, 108, 12, 11, 14, 13, 15, 10, 15, 12, 10, 14, 13, 15, 10]
```

```python
sns.boxplot(dataset)
```

```python
np.mean(dataset)
```

```python
np.median(dataset)
```

```python
statistics.mode(dataset)
```

```python
std = np.std(dataset)
print(std)
```

```python
# Finding outliers by Zscore method
outliers = []
def detect_outliers(data):
    threshold = 3  # mean 3 standard deviation
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z_score = (i - mean)/std
        print(z_score)
        if np.abs(z_score) > threshold: # abs mean roundoff value of z mean Removes the negative sign, so you only care about the magnitude.
            outliers.append(i)
    return outliers
detect_outliers(dataset)
```

```python
# By IQR Method
# 1. sort the data
# 2. calculate Q1 and Q3
# 3. callculate IQR(Q1 - Q3)
# 4. Find the lower fence(Q1 - 1.5(IQR))
# 5. Find the higher fence(Q3 + 1.5(IQR))
```

```python
dataset = sorted(dataset)
dataset
```

```python
q1 , q3 = np.percentile(dataset, [25 , 75])
print(q1 , q3)
```

```python
# Finding IQR
iqr = q3 - q1
print(iqr)
```

```python
# Finding the lower and higher fence
lower_fence = q1-(1.5*iqr)
higher_fence = q3+(1.5*iqr)
print(lower_fence , higher_fence)
```

```python
for i in dataset:
    if i > 7.5 and i < 19.5:
        print(i , end = ' ')
```

```python
print("Outliers using IQR method:")
for i in dataset:
    if i <= 7.5 or i >= 19.5:
        print(i , end = ' ')
```
